```python
import requests
import time
import pandas as pd
```

```python
def api_url(symbol, period, endTime, limit, type = 'spot', t = 'history-candles'):
    return f"https://api.bitget.com/api/{type}/v1/market/{t}?symbol={symbol}&period={period}&endTime={endTime}&limit={limit}"
```

```python
def current_milli_time():
    return round(time.time() * 1000)
current_milli_time()
```

```python
import requests
import time

boucle = True

df = pd.DataFrame()

date = current_milli_time()

limit = 200

while boucle:

    response = requests.get(api_url('RNDRUSDT_SPBL', '5min', date, limit, 'mix'))

    data = response.json()['data']
    
    if data == []:
        boucle = False
    else:
        dft = pd.DataFrame(data)
        df = pd.concat([df, dft])

        date = date - (limit * 5 * 60 * 1000)
    
# df.set_index('ts', inplace=True)
# 
# df.index = pd.to_datetime(df.index.astype(int), unit='ms', origin='unix')


# df.to_csv('futures.csv')
df
```

```python
df1 = df.copy()
    
df1.rename(columns={'ts': 'Ts', 'open': 'Open', 'close': 'Close', 'high': 'High', 'low': 'Low', 'usdtVol': 'Volume', 'baseVol': 'BaseVol'}, inplace=True)

df1['Ts'] = df1['Ts'].astype(str).apply(lambda x: x[:-3])
df1['Ts'] = pd.to_datetime(df1['Ts'].astype(int), unit='s')
df1.set_index('Ts', inplace=True)
del df1['quoteVol']
df1
```

```python
df1 = df1.astype(float)
df1['O'] = df1['Volume'] / df1['BaseVol']
df1
```

```python
df1.rename(columns={'O': 'B/Q'}, inplace=True)
```

```python
df1.to_csv('RNDR_5min_Spot.csv')
```
